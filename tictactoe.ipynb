```python
def print_board():
    for i in range(0, 9, 3):
        print("| {} | {} | {} |".format(board[i], board[i+1], board[i+2]))

def player_move(icon):
    player_number = 1 if icon == 'X' else 2
    print("Player {}'s turn ({}):".format(player_number, icon))
    while True:
        try:
            choice = int(input("Enter your move (1-9): ").strip())
            if 1 <= choice <= 9 and board[choice - 1] == " ":
                board[choice - 1] = icon
                break
            else:
                print("Invalid move. Please try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def check_win(icon):
    win_combinations = [
        [0, 1, 2], [3, 4, 5], [6, 7, 8],  # Rows
        [0, 3, 6], [1, 4, 7], [2, 5, 8],  # Columns
        [0, 4, 8], [2, 4, 6]              # Diagonals
    ]
    for combination in win_combinations:
        if all(board[i] == icon for i in combination):
            return True
    return False

def is_draw():
    return " " not in board

def initialize_board():
    return [" " for _ in range(9)]

def play_game():
    while True:
        print_board()
        player_move("X")
        if check_win("X"):
            print_board()
            print("Player 1 wins! Congratulations!")
            break
        elif is_draw():
            print_board()
            print("It's a draw!")
            break
        print_board()
        player_move("O")
        if check_win("O"):
            print_board()
            print("Player 2 wins! Congratulations!")
            break
        elif is_draw():
            print_board()
            print("It's a draw!")
            break

if __name__ == "__main__":
    board = initialize_board()
    play_game()
```
